```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Titanic-Dataset.csv")
age = df['Age'].dropna()

gender = df['Sex'].value_counts()

plt.style.use('bmh')

fig, ax = plt.subplots(1,2,figsize=(18,7))

fig.patch.set_facecolor("white")


n,bins,patches=ax[0].hist(age,
                          bins=25,
                          color="#3B82F6",
                          edgecolor="black",
                          linewidth=0.8)

ax[0].set_title("Age Distribution of Titanic Passengers",
                fontsize=18,
                fontweight="bold")

ax[0].set_xlabel("Age (Years)",fontsize=13)
ax[0].set_ylabel("Number of Passengers",fontsize=13)

mean_age=age.mean()
ax[0].axvline(mean_age,
              color="red",
              linestyle="--",
              linewidth=2,
              label=f"Mean = {mean_age:.1f}")

median_age=age.median()
ax[0].axvline(median_age,
              color="green",
              linestyle="--",
              linewidth=2,
              label=f"Median = {median_age:.1f}")

ax[0].legend()

stats=f"""
Passengers : {len(age)}
Mean       : {mean_age:.1f}
Median     : {median_age:.1f}
Min Age    : {age.min()}
Max Age    : {age.max()}
"""

ax[0].text(0.98,
           0.97,
           stats,
           transform=ax[0].transAxes,
           fontsize=11,
           verticalalignment='top',
           horizontalalignment='right',
           bbox=dict(facecolor='white',
                     edgecolor='black',
                     boxstyle='round'))


colors=["#2563EB","#F59E0B"]

bars=ax[1].bar(gender.index,
               gender.values,
               color=colors,
               width=0.55,
               edgecolor="black")

ax[1].set_title("Gender Distribution",
                fontsize=18,
                fontweight="bold")

ax[1].set_xlabel("Gender",fontsize=13)
ax[1].set_ylabel("Passengers",fontsize=13)

total=gender.sum()

for bar in bars:

    height=bar.get_height()

    percentage=height/total*100

    ax[1].text(bar.get_x()+bar.get_width()/2,
               height+10,
               f"{height}\n({percentage:.1f}%)",
               ha='center',
               fontsize=12,
               fontweight='bold')
for a in ax:
    a.spines['top'].set_visible(False)
    a.spines['right'].set_visible(False)

plt.suptitle("Titanic Dataset - Distribution Analysis",
             fontsize=24,
             fontweight="bold",
             color="#1F2937")

plt.tight_layout(rect=[0,0,1,0.95])

plt.show()
```
